```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_input = input("Please enter the x values, separated by spaces: ")
y_input = input("Please enter the y values, separated by spaces: ")
```

```python
import numpy as np

# Converting user input strings to numpy arrays
x = np.array(input("Please enter the x values, separated by spaces: ").split(), dtype=np.float32)
y = np.array(input("Please enter the y values, separated by spaces: ").split(), dtype=np.float32)
```

```python
# Reshaping x and y arrays for TensorFlow
x = np.reshape(x, (-1, 1))
y = np.reshape(y, (-1, 1))
```

```python
import tensorflow as tf

# Defining the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])
```

```python
# Compiling the model
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# Training the model and storing the training history
history = model.fit(x, y, epochs=1000, verbose=0)

```

```python
# Making predictions
y_pred = model.predict(x)
```

```python
# Printing the predictions
for i in range(len(x)):
    print(f"For x = {x[i][0]}, predicted y = {y_pred[i][0]:.2f}, actual y = {y[i][0]}")
```

```python
# Taking user input for prediction
user_input = float(input("Please enter a value of x to predict y: "))
user_pred = model.predict([[user_input]])
print(f"For x = {user_input}, predicted y = {user_pred[0][0]:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'])
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')

# Plotting the regression line and data points graph
plt.subplot(1, 2, 2)
plt.scatter(x, y, label='Actual Data')
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.title('Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.tight_layout()
plt.show()
```
